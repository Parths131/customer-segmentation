```python
df = pd.read_excel('/content/online_retail_II.xlsx', sheet_name='Year 2010-2011')
```

```python
#df = df.dropna(subset=['CustomerID'])
df = df[~df['Invoice'].astype(str).str.startswith('C')]
```

```python
df
```

```python
basket = (df[df['Quantity'] > 0]
          .groupby(['Invoice', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('Invoice'))
```

```python
basket = basket.map(lambda x: 1 if x >= 1 else 0)
```

```python
basket.head()
```

```python
item_counts = basket.sum().sort_values(ascending=False)
```

```python
item_counts
```

```python
bad_keywords = ['wrongly', 'incorrectly', 'test', 'adjust', 'credit', 'returned', 'check?', 'website fixed']
```

```python
basket_cleaned = basket.loc[:, ~basket.columns.str.lower().str.contains('|'.join(bad_keywords))]
```

```python
print(basket_cleaned.shape)
```

```python
item_counts = basket_cleaned.sum().sort_values(ascending=False)
```

```python
item_counts
```

```python
basket_cleaned
```

```python
product_counts = basket_cleaned.sum().sort_values(ascending=False)
```

```python
#from mlxtend.frequent_patterns import apriori, association_rules
```

```python
top_items = product_counts.head(50).index
```

```python
basket_small = basket_cleaned[top_items]
```
